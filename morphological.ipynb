```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

A = 255-cv2.imread('D:\HCMUTE\BTVN-XuLyAnh\Image\\testpic\component.bmp', 0)
_, A = cv2.threshold(A, 0, 1, cv2.THRESH_BINARY)
plt.imshow(A, cmap='gray')
plt.show()
Label = np.zeros(A.shape)
C = np.array([[341, 73, 593, 1185, 1661, 413, 869, 1633, 237, 605, 1037, 1481, 453, 901, 1321, 261, 653, 1013, 1537, 1481], [457, 753, 773, 481, 165, 1017, 1117, 913, 1453, 1561, 1509, 1381, 1973, 1773, 1777, 2293, 2245, 2149, 2029, 2401]]).transpose()
B = np.ones((5, 5), 'uint8')
for i in range(len(C)):
    X1 = np.zeros(A.shape, 'uint8')
    X0 = A.copy()
    X1[C[i, 0], C[i, 1]] = 1
    while (np.sum(X0 != X1) > 0):
        X0 = X1.copy()
        X1 = cv2.dilate(X0, B, iterations=1) & A
        #plt.imshow(X1,cmap='gray'); plt.show()
    #plt.imshow(X1, cmap='gray')
    #plt.show()
    Label[X1 == 1] = i + 1
    print('Component ', i + 1, " co so diem anh la: ", np.sum(Label == i + 1))

fig = plt.figure(dpi=300)
plt.subplot(1, 2, 1), plt.xticks([])
plt.yticks([])
plt.imshow(A, 'gray'), plt.title('Input')
plt.subplot(1, 2, 2), plt.xticks([]), plt.yticks([])
plt.imshow(Label*100, 'gray'), plt.title('All components')
plt.tight_layout()
```
